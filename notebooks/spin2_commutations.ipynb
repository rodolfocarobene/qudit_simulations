```python
def commute(a, b):
    c = a@b
    d = b@a
    return (np.array((c-d)) == np.zeros(c.shape)).all()
```

```python
from scipy.linalg import ishermitian as ish

def check_matrix_properties(matrix):
    matrix = np.array(matrix, dtype=complex)
    is_hermitian = ish(matrix)
    identity_matrix = np.eye(matrix.shape[0])
    is_self_inverse = np.allclose(matrix @ matrix, identity_matrix)
    is_anti_symmetric = np.allclose(matrix, -matrix.T)    
    return {
        "is_hermitian": is_hermitian,
        "is_self_inverse": is_self_inverse,
        "is_anti_symmetric": is_anti_symmetric
    }
```
