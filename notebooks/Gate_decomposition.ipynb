## Spinless Hamiltonian gates

**Potential Hamiltonian**

$H = \frac{1}{4} (I-\tilde\Gamma)_1(I-\tilde\Gamma)_2$

```python
H = (1/4) * TensorProduct(sy_id - sy_gamma_5, sy_id) * TensorProduct(sy_id, sy_id - sy_gamma_5)
mat_H = np.array(H, dtype=np.complex128)

H
```

```python
tau = 1
evol_H = Matrix(np.round(expm(-1j * tau * mat_H), 3))
evol_H
```

```python
tau = 1 #np.pi
x_rot_a = Z_P_ij(0, 1, tau)._unitary_() @ Z_P_ij(2, 3, -tau)._unitary_()


l = TensorProduct(x_rot_a, x_rot_a)
l = sy_ucsum * l * sy_ad_ucsum
l = np.round(np.array(l, dtype=np.complex128), 3)

Matrix(l)
```

**Kinetic Hamiltonian**

$H = \frac{1}{4} (I-\tilde\Gamma)_1(I-\tilde\Gamma)_2$

## Example

```python
h1 = -1j * TensorProduct(sy_gamma_2 * sy_gamma_5, sy_gamma_1)
mat_h1 = np.array(h1, dtype=np.complex128)
ev_h1 = Matrix(np.round(expm(-1j * 0.01 * mat_h1), 3))
ev_h1
```

```python
tau = 0.01 * 2
x_rot_a = X_P_ij(0, 2, +tau)._unitary_()
x_rot_b = X_P_ij(1, 3, -tau)._unitary_()

l = sy_ucsum * TensorProduct(x_rot_a @ x_rot_b, sy_id) * sy_ad_ucsum
l = np.round(np.array(l, dtype=np.complex128), 3)

Matrix(l)
```
